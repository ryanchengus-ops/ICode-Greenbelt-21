```python
import matplotlib.pyplot as plt
import scipy.stats as stats # Changed import to directly get stats submodule

x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]
y = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

slope, intercept, r, p, std_err = stats.linregress(x, y) # Called linregress as a function
def myfunc(x_val):
  return slope * x_val + intercept
mymodel = list(map(myfunc, x))
plt.plot(x,mymodel)
plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

x = np.random.normal(5.0, 1.0, 1000)
y = np.random.normal(10.0, 2.0, 1000)
plt.scatter(x, y)
plt.show()
slope, intercept, r, p, std_err = stats.linregress(x, y) # Called linregress as a function
def myfunc(x_val):
  return slope * x_val + intercept
mymodel = list(map(myfunc, x))
plt.plot(x,mymodel)
plt.scatter(x, y)
plt.show()
```
